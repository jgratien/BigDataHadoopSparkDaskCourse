```python
from dask import delayed
from dask.distributed import LocalCluster, Client
import time

@delayed
def inc(x):
    time.sleep(10)
    return x+1

@delayed
def add(x, y):
    time.sleep(11)
    return x + y
```

```python
cluster = LocalCluster()
client = Client(cluster)
```

```python
x = inc(15)
y = inc(30)

total = add(x,y)

print(total)
```

```python
total.visualize()
```
